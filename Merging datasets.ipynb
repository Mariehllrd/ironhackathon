### Importing librairies

```python
import numpy as np
import pandas as pd
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

### Importing files

```python
paygap_07 = pd.read_csv('./data/paygap_07.csv')
paygap_07 = paygap_07.drop('Unnamed: 0', axis = 1)

paygap_08 = pd.read_csv('./data/paygap_08.csv')
paygap_08 = paygap_08.drop('Unnamed: 0', axis = 1)

paygap_09 = pd.read_csv('./data/paygap_09.csv')
paygap_09 = paygap_09.drop('Unnamed: 0', axis = 1)

paygap_10 = pd.read_csv('./data/paygap_10.csv')
paygap_10 = paygap_10.drop('Unnamed: 0', axis = 1)

paygap_11 = pd.read_csv('./data/paygap_11.csv')
paygap_11 = paygap_11.drop('Unnamed: 0', axis = 1)

paygap_12 = pd.read_csv('./data/paygap_12.csv')
paygap_12 = paygap_12.drop('Unnamed: 0', axis = 1)

paygap_13 = pd.read_csv('./data/ds_13.csv')
paygap_13 = paygap_13.drop('Unnamed: 0', axis = 1)

paygap_14 = pd.read_csv('./data/ds_14.csv')
paygap_14 = paygap_14.drop('Unnamed: 0', axis = 1)

paygap_15 = pd.read_csv('./data/ds_15.csv')
paygap_15 = paygap_15.drop('Unnamed: 0', axis = 1)

paygap_16 = pd.read_csv('./data/ds_16.csv')
paygap_16 = paygap_16.drop('Unnamed: 0', axis = 1)

paygap_17 = pd.read_csv('./data/ds_17.csv')
paygap_17 = paygap_17.drop('Unnamed: 0', axis = 1)
```

```python
paygap_13['Year'] = 2013
paygap_14['Year'] = 2014
paygap_15['Year'] = 2015
paygap_16['Year'] = 2016
paygap_17['Year'] = 2017
```

```python
paygap_13 = paygap_13.rename(columns = {'2012':'Total'})
paygap_14 = paygap_14.rename(columns={'2011':'Total'})
paygap_15 = paygap_15.rename(columns={'2010':'Total'})
paygap_16 = paygap_16.rename(columns={'2009':'Total'})
paygap_17 = paygap_17.rename(columns={'2008':'Total'})
```

### Concatenate datasets

```python
df = pd.concat([paygap_07, paygap_08, paygap_09, paygap_10, paygap_11, paygap_12, paygap_13, paygap_14, paygap_15,
               paygap_16, paygap_17], sort = True)
```

```python
# Dropping nan
df = df.replace(0,np.nan)
df = df.dropna(axis = 'columns', how = 'all')
```

```python
df[['Administration', 'Arts and Sports', 'Construction', 
       'Education', 'Electricity and Gas', 'Extractive Ind',
       'Finance and Insurance', 'Hospitality', 'Information and Communication',
       'Manufacturing', 'Real Estate', 'Scientific Consultancy', 'Social Work',
       'Total', 'Transportation and Storage', 'Water and Sewage', 'Wholesale']] = df[['Administration', 'Arts and Sports', 'Construction', 
       'Education', 'Electricity and Gas', 'Extractive Ind',
       'Finance and Insurance', 'Hospitality', 'Information and Communication',
       'Manufacturing', 'Real Estate', 'Scientific Consultancy', 'Social Work',
       'Total', 'Transportation and Storage', 'Water and Sewage', 'Wholesale']].astype('float')
```

```python
df['Year'] = df['Year'].astype('O')
```

```python
df.Total.fillna(df.mean(axis=1), inplace = True)
```

```python
df = df.sort_values('Country')
```

```python
df.isna().sum()
```

```python
df = df.dropna(axis = 'index', thresh = 6)
df.columns
```
