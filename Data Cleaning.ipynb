```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = None
pd.options.display.max_rows = None

from googletrans import Translator
```

# 2007

```python
# Importing dataset
paygap_07 = pd.read_excel('./data/2008.xlsx', header = [10])
```

```python
# Dropping nulls
paygap_07 = paygap_07.dropna(axis = 1, how = 'all')
paygap_07 = paygap_07.dropna(axis = 0, how = 'any')
```

```python
# Renaming columns
to_rename = ['Unnamed: 0', 'Agricultura, floresta e pescas', 'Indústrias extractivas', 'Indústrias transformadoras', 
             'Abastecimento de electricidade, gás, vapor e ar condicionado', 
             'Abastecimento de água, rede de esgotos, ...', 'Construção', 'Comércio por grosso e a retalho, ...', 
             'Transportes e armazenagem', 'Alojamento, restauração e similares', 'Informação e comunicação',
             'Actividades financeiras e de seguros', 'Actividades imobiliárias', 
             'Actividades de consultoria, científicas e técnicas', 
             'Actividades administrativas e dos serviços de apoio', 
             'Administração Pública e defesa; segurança social obrigatória', 'Educação',
             'Actividades de saúde humana e apoio social',
             'Actividades artísticas, de espectáculos, desportivas e recreativas',
             'Outras actividades de serviços', 'Actividades dos agregados domésticos como empregadores, ...',
             'Actividades de organizações e entidades extraterritoriais', 'Desconhecido']

new_columns = ['Country',"Agriculture","Extractive Ind", "Manufacturing", "Electricity and Gas",
               "Water and Sewage","Construction", "Wholesale", "Transportation and Storage","Hospitality",
               "Information and Communication", "Finance and Insurance", "Real Estate", "Scientific Consultancy",
               "Administration", "Public Sector", "Education","Social Work", "Arts and Sports", 
               "Other Services", "Domestic Work", "International Administration", "Unknown"]

colnames = dict(zip(to_rename, new_columns))
paygap_07 = paygap_07[['Unnamed: 0', 'Agricultura, floresta e pescas', 'Indústrias extractivas', 
                       'Indústrias transformadoras', 
             'Abastecimento de electricidade, gás, vapor e ar condicionado', 
             'Abastecimento de água, rede de esgotos, ...', 'Construção', 'Comércio por grosso e a retalho, ...', 
             'Transportes e armazenagem', 'Alojamento, restauração e similares', 'Informação e comunicação',
             'Actividades financeiras e de seguros', 'Actividades imobiliárias', 
             'Actividades de consultoria, científicas e técnicas', 
             'Actividades administrativas e dos serviços de apoio', 
             'Administração Pública e defesa; segurança social obrigatória', 'Educação',
             'Actividades de saúde humana e apoio social',
             'Actividades artísticas, de espectáculos, desportivas e recreativas',
             'Outras actividades de serviços', 'Actividades dos agregados domésticos como empregadores, ...',
             'Actividades de organizações e entidades extraterritoriais', 'Desconhecido']]

paygap_07 = paygap_07.rename(columns = colnames)
```

```python
# Keeping only country code
paygap_07['Country'] = paygap_07['Country'].apply(lambda x: x.split('-')[-1].strip())

translator = Translator()
paygap_07['Country'] = paygap_07['Country'].apply(Translator().translate, src='pt', dest='en').apply(getattr, args=('text',))
paygap_07['Country'] = paygap_07['Country'].replace('low Country','Netherlands').replace('malt','Malta').str.capitalize()
```

```python
# Dropping first two rows
paygap_07 = paygap_07.drop([0, 1])
```

```python
# Dropping nan
paygap_07 = paygap_07.replace(0,np.nan)
paygap_07 = paygap_07.dropna(axis = 1, how = 'all')
paygap_07 = paygap_07.dropna(axis = 0, how = 'any')
```

```python
# Adding year column
paygap_07['Year'] = 2007
```

```python
# Saving clean dataset
paygap_07.to_csv('./data/paygap_07.csv')
```
